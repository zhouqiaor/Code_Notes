```python
import numpy as np
```

```python
matrix_load = np.loadtxt('Example.csv', dtype=np.int, delimiter=",")
```

```python
matrix_load_0429 = np.loadtxt('Example_0429.csv', dtype=np.int, delimiter=",")
```

```python
import networkx as nx
```

```python
nRow, nCol = matrix_load.shape
A = np.zeros((nRow+nCol, nRow+nCol))
A[:nRow, nRow:] = matrix_load[:,:]
A[nRow:, :nRow] = matrix_load.T[:,:]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.matshow(matrix_load)
```

```python
plt.matshow(A)
plt.savefig('A.tif', dpi=400)
```

```python
# graph
G = nx.from_numpy_matrix(A, create_using=nx.Graph)
```

```python
len(list(nx.bridges(G)))
```

```python
nx.has_bridges(G)
```

```python
nx.is_chordal(G)
```
